```python
import matplotlib.pyplot as plt

# Datos
labels = ['Padre que perdió el trabajo', 'Emocionalmente abusado por un padre', 'Fisicamente abusado por un padre', 'Otros']
sizes = [29, 55, 11, 5]  # Porcentajes de cada categoría
colors = ['#ff9999', '#ff6666', '#ff3333', '#ff0000']
explode = (0, 0, 0, 0)  # 'Explota' el primer slice

# Crear el gráfico de torta
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')  # Igualar los ejes para que sea un círculo perfecto
plt.title('Distribución de Categorías')
plt.show()
```
