```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV
file_path = '/content/nics-latest.csv'
df = pd.read_csv(file_path)

# Filtrar los datos para Ohio en 2021
ohio_2021 = df[(df['state'] == 'Ohio') & (df['year'] == 2021)]

# Extraer las columnas relevantes
handgun_sales = ohio_2021['handgun_adjusted']
long_gun_sales = ohio_2021['long_gun_adjusted']
total_sales = ohio_2021['total_adjusted']

# Crear la visualización
plt.figure(figsize=(10, 6))

# Configurar los datos para la visualización
bar_width = 0.3
index = ohio_2021.index

# Graficar las ventas ajustadas de armas de mano y rifles largos
plt.bar(index - bar_width, handgun_sales, width=bar_width, label='Handgun Adjusted')
plt.bar(index, long_gun_sales, width=bar_width, label='Long Gun Adjusted')
plt.bar(index + bar_width, total_sales, width=bar_width, label='Total Adjusted')

# Personalizar el gráfico
plt.xlabel('Mes')
plt.ylabel('Cantidad de armas vendidas ajustadas')
plt.title('Ventas ajustadas de armas en Ohio en 2021')
plt.xticks(index, ohio_2021['month'], rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV
file_path = '/content/nics-latest.csv'
df = pd.read_csv(file_path)

# Filtrar los datos para el año 2020 en todos los estados
year_2020 = df[df['year'] == 2020]

# Sumar las ventas ajustadas por categoría en todo el país
handgun_total = year_2020['handgun_adjusted'].sum()
long_gun_total = year_2020['long_gun_adjusted'].sum()
total_sales = year_2020['total_adjusted'].sum()

# Etiquetas para las barras
categories = ['Handgun Adjusted', 'Long Gun Adjusted', 'Total Adjusted']
values = [handgun_total, long_gun_total, total_sales]

# Crear la visualización
plt.figure(figsize=(10, 6))

# Graficar las ventas ajustadas de armas de mano, rifles largos y el total
plt.bar(categories, values, color=['blue', 'green', 'orange'])

# Personalizar el gráfico
plt.xlabel('Categoría de Arma')
plt.ylabel('Cantidad de armas vendidas ajustadas')
plt.title('Ventas ajustadas de armas en Estados Unidos en 2020')
plt.grid(True)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
import pandas as pd
from prettytable import PrettyTable

# Cargar el archivo CSV
file_path = '/content/nics-latest.csv'
df = pd.read_csv(file_path)

# Filtrar los datos para el año 2020
year_2020 = df[df['year'] == 2020]

# Agrupar por estado y sumar las ventas ajustadas de armas de mano y rifles largos
state_totals = year_2020.groupby('state')[['handgun_adjusted', 'long_gun_adjusted']].sum()

# Crear una tabla bonita (PrettyTable)
table = PrettyTable(['Estado', 'Handgun Adjusted', 'Long Gun Adjusted'])

# Iterar sobre cada fila de state_totals y agregar a la tabla
for index, row in state_totals.iterrows():
    table.add_row([index, row['handgun_adjusted'], row['long_gun_adjusted']])

# Alinear los datos a la izquierda
table.align = 'l'

# Mostrar la tabla
print(table)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Cargar el archivo CSV
file_path = '/content/nics-latest.csv'
df = pd.read_csv(file_path)

# Filtrar los datos para el año 2020 y excluir 'United States' de 'state'
year_2020 = df[(df['year'] == 2020) & (df['state'] != 'United States')]

# Agrupar por estado y sumar las ventas ajustadas de armas de mano y rifles largos
state_totals = year_2020.groupby('state')[['handgun_adjusted', 'long_gun_adjusted']].sum()

# Ordenar por la suma de ventas de armas de mano para una mejor visualización
state_totals = state_totals.sort_values(by='handgun_adjusted', ascending=False)

# Definir colores claros y oscuros
color_lightblue = '#77ACF1'  # Azul claro
color_darkblue = '#2D70A3'   # Azul oscuro

# Crear la visualización (gráfico de barras apiladas)
plt.figure(figsize=(12, 8))

# Graficar las ventas ajustadas de armas de mano y rifles largos por estado
bars_handgun = plt.bar(state_totals.index, state_totals['handgun_adjusted'], color=color_lightblue, label='Handgun Adjusted')
bars_longgun = plt.bar(state_totals.index, state_totals['long_gun_adjusted'], bottom=state_totals['handgun_adjusted'], color=color_darkblue, label='Long Gun Adjusted')

# Personalizar el gráfico
plt.xlabel('Estado')
plt.ylabel('Cantidad de armas vendidas ajustadas')
plt.title('Ventas ajustadas de armas de mano y rifles largos por estado en 2020 (Excluyendo Estados Unidos)')
plt.xticks(rotation=45)
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Cargar el archivo CSV
file_path = '/content/nics-latest.csv'
df = pd.read_csv(file_path)

# Filtrar los datos para el año 2020 y excluir 'United States' de 'state'
year_2020 = df[(df['year'] == 2020) & (df['state'] != 'United States')]

# Agrupar por estado y sumar las ventas ajustadas de armas de mano y rifles largos
state_totals = year_2020.groupby('state')[['handgun_adjusted', 'long_gun_adjusted']].sum()

# Ordenar por la suma de ventas de armas de mano para una mejor visualización
state_totals = state_totals.sort_values(by='handgun_adjusted', ascending=False)

# Seleccionar los 10 estados con mayores ventas de armas de mano
top_10_states = state_totals.head(10)

# Definir colores claros y oscuros
color_lightblue = '#77ACF1'  # Azul claro
color_darkblue = '#2D70A3'   # Azul oscuro

# Crear la visualización (gráfico de barras apiladas)
plt.figure(figsize=(12, 8))

# Graficar las ventas ajustadas de armas de mano y rifles largos por estado
bars_handgun = plt.bar(top_10_states.index, top_10_states['handgun_adjusted'], color=color_lightblue, label='Handgun Adjusted')
bars_longgun = plt.bar(top_10_states.index, top_10_states['long_gun_adjusted'], bottom=top_10_states['handgun_adjusted'], color=color_darkblue, label='Long Gun Adjusted')

# Personalizar el gráfico
plt.xlabel('Estado')
plt.ylabel('Cantidad de armas vendidas ajustadas')
plt.title('Top 10 estados con mayores ventas ajustadas de armas de mano y rifles largos en 2020')
plt.xticks(rotation=45)
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()

plt.savefig('top_10_states_sales.jpg')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV
file_path = '/content/nics-latest.csv'
df = pd.read_csv(file_path)

# Filtrar los datos para el año 2020 y excluir 'United States' de 'state'
year_2020 = df[(df['year'] == 2020) & (df['state'] != 'United States')]

# Agrupar por estado y sumar las ventas ajustadas de armas de mano y rifles largos
state_totals = year_2020.groupby('state')[['handgun_adjusted', 'long_gun_adjusted']].sum()

# Ordenar por la suma de ventas de armas de mano para una mejor visualización
state_totals = state_totals.sort_values(by='handgun_adjusted', ascending=False)

# Seleccionar los 10 estados con mayores ventas de armas de mano
top_10_states = state_totals.head(10)

# Calcular los totales de ventas de armas de mano y rifles largos a nivel nacional
total_handgun = top_10_states['handgun_adjusted'].sum()
total_longgun = top_10_states['long_gun_adjusted'].sum()

# Crear etiquetas para el gráfico circular
labels = ['Handgun Adjusted', 'Long Gun Adjusted']

# Datos para el gráfico circular (porcentajes)
sizes = [total_handgun, total_longgun]

# Colores para cada sección del gráfico
colors = ['#77ACF1', '#2D70A3']  # Azul claro y azul oscuro

# Crear la figura y el gráfico circular
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Personalizar el gráfico
plt.title('Porcentaje de venta')
```
